Write a function to generate simulated synthetic panel data where conditioning on covariates is real important. We allow for selection into treatment based on time varying covariates, time fixed effects, and unit fixed effects.

```python
def synth_panel(seed_number = 2123, 
                time_units = 10, 
                x_units = 100, 
                K_dim = 5 ,
               at = 2):
    '''
    seed_number  random_number_set
    time_units   how many time periods?
    x_units      how many units?
    K_dim        how many covariates (assume uniform)
    at           ATET estimate
    '''
    def censor_at_x(x,cr):
        x[x >=   cr]   =    cr
        x[x <=-1*cr] = -1*cr
        return x
    censorvalue = 1

    
    ## Set the seed for random number
    xfun = np.random.RandomState(seed_number)
    
    ## Construct covariates, time-invariant fixed effects, and first period outcome
    X = censor_at_x( xfun.normal(0,2, size= (x_units, K_dim) ), 5  )  
    X_coef    = xfun.uniform(low=-2, high=2, size = K_dim)
    X_coef_W  = xfun.uniform(low= 2, high=2, size = K_dim)
    x_list = ['x'+str(k) for k in range(K_dim)]
        
    ## Unit fixed effects
    fe = censor_at_x( xfun.normal(0,1, size=x_units), censorvalue)
    ## Time fixed effects
    te = xfun.normal(0,0.5)
    
    ## First period outcome, where the outcome is not a function of the covariates.
    ## Also use this opportunity to determine treatment status
    output_df = pd.DataFrame(data=X, columns=x_list)
    output_df['Y'] = fe + censor_at_x( xfun.normal(0,0.5, size=x_units), censorvalue)
    output_df['T'] = pd.to_datetime('2020/01/01')
    output_df['unitid'] =  np.arange(x_units).astype(int)

    ## Calculate treatment status
    latent_treatment= np.dot(output_df[x_list], X_coef_W) +\
                    xfun.normal(0,1,size=x_units)
    ## First define treated units based on the latent treatment exceeding some threshold    
    threshold_lower = np.percentile( latent_treatment, q=75)
    threshold_upper = np.percentile( latent_treatment, q=90)    
    
    ids = list(output_df.loc[(threshold_lower<=latent_treatment) &\
                             (latent_treatment < threshold_upper)\
                                                  ]['unitid'].unique())

    ids = list( output_df.loc[(latent_treatment>= threshold_upper) ]['unitid'].unique() )

    treated_units = (output_df['unitid'].isin(ids))            
    output_df['treatment'] = (treated_units)*1
    
    ## Decay parameter
    decay = 0.80
    covariates= [r for r in output_df.columns if 'x' in r ]
    
    ## Create the panel by setting up an AR(1) type function.
    ## Also allow a time fixed effect and put the covariates into the outcome
    for t in range(1,time_units):    
        
        prev_date = (pd.to_datetime('2020/01/01')+pd.offsets.DateOffset(months=t-1))
        prev_data = decay*( output_df.loc[(output_df['T']==prev_date)][covariates]     )
        ## perturb it a little with AR(1)
        prev_data += (1-decay)*censor_at_x( xfun.normal(0,1, size=(x_units,K_dim)), censorvalue)
                
        ## Time fixed effect, which is correlated over time
        if t==1:
            te = xfun.normal(0,1)
        else:
            te = decay*te + xfun.normal(0,1)
        
        prev_data['Y'] = decay*np.dot(prev_data,X_coef) + fe + te + censor_at_x( xfun.normal(0,0.5, size=x_units), censorvalue)
        prev_data['T'] = pd.to_datetime('2020/01/01') + pd.offsets.DateOffset(months=t)
        prev_data['unitid'] = np.arange(x_units).astype(int)
        
        treated_units = (prev_data['unitid'].isin(ids))            
        prev_data['treatment'] = (treated_units)*1

        output_df = pd.concat([ output_df, prev_data])

    output_df['Y_control'] = output_df['Y'].values
    
    ## Assign treatment values at treatment times
    ## Let treatment value depend on the covariates over time. Use a different
    ## set of parameters to aggregate the covariates so treatment status is not
    ## as tightly correlated with the control outcome.
    ## Let treatment all happen at the same time.
        
    ## Assign treatment time
    treatment_time = (output_df['T'] >= pd.to_datetime('2020/01/01')+\
                      pd.offsets.DateOffset(months=int(time_units*0.75)-1))         

    ## Treatment unit assignment
    ## Treatment ATET ground truth for treated units
    output_df['treatment_GT'] = (treatment_time & (output_df['treatment']==1) )*at

    ## Treatment interaction post and unit
    output_df['treatmentxpost'] = (treatment_time &  (output_df['treatment']==1) )*1
    
    ## Apply treatment to outcome
    output_df['Y'] += (treatment_time &  (output_df['treatment']==1)  )*at


    output_df['unitid'] = 'unit' + output_df['unitid'].apply(str)
    return output_df
```

## Basecase: Standard TWFE and IPW (via Abadie, 2005 ) Models

Let the standard two-way fixed effects (TWFE) and inverse propensity score (IPW) following Abadie, 2005 be:

$$TWFE: Y_{it} = \alpha_i + \gamma_t + \tau W_{it} + \epsilon_{it}$$
for unit fixed effects $\alpha_i$, time fixed effects $\gamma_t$, and the post-treatment indicator $W_{it}$ with idiosyncratic errors $\epsilon_{it}$. $\tau$ is our ATET estimate.

The IPW model is:
$$\tau = \dfrac{1}{\mathbb{E}[D_{i}]}  \mathbb{E}\Big[  \dfrac{ D_{i} - \hat{D_{i}} }{  1 - \hat{D_{i}}  } \big( Y^1_{i} - Y^0_{i} \big) \Big] $$

Where $\hat{D_{i}}$ is the propensity score for $D_{i}$ based on time-varying covariates $X_{it}$ and $Y^1$ is the observed outcome of all units post-treatment time, and $Y^0$ is the observed outcome all units pre-treatment time.

The constrast between unit-level variation in $D_i$ and time-varying information in $Y^1$ and $Y^0$ gets us the difference-in-difference estimate. We can see this by recognizing that the weight is essentially equal to 1 when $D_i=1$.

We can see this by differentiating between four scenarios, for $Y_{d,t}$:
1. $D_i = 1, T_i=1: \dfrac{1}{\mathbb{E}[D_{i}]} Y_{1,1} $
2. $D_i = 0, T_i=1: \dfrac{-\hat{D_i}}{(1-\hat{D_i})\mathbb{E}[D_{i}]} Y_{0,1} $
3. $D_i = 1, T_i=0: \dfrac{1}{\mathbb{E}[D_{i}]} Y_{1,0} $
4. $D_i = 0, T_i=0: \dfrac{-\hat{D_i}}{(1-\hat{D_i})\mathbb{E}[D_{i}]} Y_{0,0} $

Then we can re-write the IPW model as the post-pre difference among treated ($D_i=1$) and post-pre difference among control ($D_i=0$):
$$ \dfrac{1}{\mathbb{E}[D_{i}]}(Y_{1,1} - Y_{1,0}) - \dfrac{-\hat{D_i}}{(1-\hat{D_i})\mathbb{E}[D_{i}]}(Y_{0,1} - Y_{0,0}) $$



```python
def DiD(data, Y, unit_column, time_column, post):
    '''
    data            dataframe with all the data features and goodies we need
    Y              string name of the outcome
    unit_column    name of the unit id column for generating fixed effects
    time_column    name of the time column for generating fixed effects
    post           list of names of post interaction indicators
    '''
    unit_fe = pd.get_dummies(data[unit_column], drop_first=True, dtype=float)
    time_fe = pd.get_dummies(data[time_column], drop_first=True, dtype=float)
    X = sm.add_constant( pd.concat([data[post], unit_fe, time_fe], axis=1) )
    ols_fitted = sm.OLS(data[Y],X).fit()
    return ols_fitted

```

```python
def ipw_did(data, Y, unit_column, time_column, treatment, post,
           covariates, propensity_model):
    unit_fe = pd.get_dummies(data[unit_column], drop_first=True, dtype=float)
    time_fe = pd.get_dummies(data[time_column], drop_first=True, dtype=float)

    ## Find out the treatment time
    treatment_time = data.loc[(data[post]==1)][time_column].min()
    treated_units = data.loc[(data[treatment]==1)][unit_column].unique()
        
    X = pd.concat([data[covariates], unit_fe, time_fe], axis=1)
     
    ## Train a propensity score model on the treatment 
    p_fitted = propensity_model.fit(X,
                                    data[treatment] )

    ## Only predict for the control and post-treatment samples    
    p_hat = p_fitted.predict_proba(X)[:,1]
        
    numerator = (data[treatment] - p_hat)
    denominator = (1-p_hat)
#     denominator = (1-p_hat)*p_hat
    
    weight = numerator/denominator   
    
    y_11 = data.loc[(data[treatment]==1) & (data[time_column] >= treatment_time) ][Y]
    y_10 = data.loc[(data[treatment]==1) & (data[time_column] < treatment_time) ][Y]
    y_01 = data.loc[(data[treatment]==0) & (data[time_column] >= treatment_time) ][Y]
    y_00 = data.loc[(data[treatment]==0) & (data[time_column] < treatment_time) ][Y]    

    diff_treated = y_11.mean() - y_10.mean()
    
    weight_01 = weight[(data[treatment]==0) & (data[time_column] >= treatment_time) ]
    weight_00 = weight[(data[treatment]==0) & (data[time_column] < treatment_time) ]
    diff_control = (y_01*weight_01).mean() -  (y_00*weight_00).mean()
    
    diff = diff_treated - diff_control

    return {'atet':diff,'weight':weight,
            'phat':p_hat}

```

# Proposed DRDiD Model
Before going into doubly robust DiD, let's first review the regression adjustment portion by itself.
Note that DiD estimate the ATET, which is:
$$\mathbb{E}[Y_1(1) - Y_1(0) | D = 1]$$
For $Y_D(D^*)$ for a unit in group $D$ that is actually treated with $D^*$, where $D$ denotes treatment group, which are treated after the treatment time.
Through law of iterated expectations, we can re-write the above as:
$$\mathbb{E}[Y_1(1)|D=1] - \mathbb{E}[Y_1(0)|D=1] = \mathbb{E}[Y_1|D=1] - \mathbb{E}[Y_1(0)|D=1] $$

Note that we can observe $\mathbb{E}[Y_1|D=1] $ directly, so it's a matter of predicting $\mathbb{E}[Y_1(0)|D=1]$.

This ties into the regression adjustment specification, which is inherited from difference-in-differences. The first difference is: $Y_{1,1} - Y_{1,0}$, which is the pre-post difference for the treatment group, and $Y_{0,1} - Y_{0,0}$ which is the pre-post difference for the control group. The regression adjustment expands on the pre-post difference for the control group by just removing more noise in the control outcomes:
$$\big(\mathbb{E}[Y_{1,1}] - \mathbb{E}[Y_{1,0}] \big) - \big(  \mathbb{E}_1[\hat{Y}_{0,1}] - \mathbb{E}_1[\hat{Y}_{0,0}]\big)$$
Where the expectation $\mathbb{E}_1$ is over treatment units. So that's a prediction of how the treatment units would perform if they were in fact control.

We now focus on the panel version of Sant'Anna and Zhao (2018)'s doubly robust DiD (DRDiD).
$$\tau_{DRDiD} = \mathbb{E}[A-B]\times C$$
where:
\begin{align}
A = & \dfrac{D_i}{ \mathbb{E}[D_i]  } \\
B = & \dfrac{g(X)(1-D_i)}{1-g(X)} \times \mathbb{E}^{-1}[ \dfrac{g(X)(1-D_i)}{1-g(X)}] \\
C = & (Y^1  -  Y^0) - (\mu_{0,1}(X)) - \mu_{0,0}(X))
\end{align}
so fully written out:
$\tau_{DRDiD} = \mathbb{E}\Big[  \big(\dfrac{D_i}{ \mathbb{E}[D_i]  }   -  \dfrac{g(X)(1-D_i)}{1-g(X)} \times \mathbb{E}^{-1}[ \dfrac{g(X)(1-D_i)}{1-g(X)}] \big) \big( (Y^1  -  Y^0) - (\mu_{0,1}(X)) - \mu_{0,0}(X)) \big)   \Big]$

where:
1. $D_i$ is the treatment indicator that is 1 for units that experience treatment in the post-period and 0 otherwise, 
2. $Y^1$ is the observed outcome of units post-treatment,
3. $Y^0$ is the observed outcome of units pre-treatment,
4. $\mu_{0,1}$ is the estimated control outcome post treatment, and 
5. $\mu_{0,0}$ is the estimated control outcome pre-treatment. 
6. $g(X)$ is predicted propensity score.

```python
def drdid(data, Y, unit_column, time_column, treatment, post,
             covariates, 
             propensity_model,
             outcome_model):
    unit_fe = pd.get_dummies(data[unit_column], drop_first=True, dtype=float)
    time_fe = pd.get_dummies(data[time_column], drop_first=True, dtype=float)
    
    X = pd.concat([data[covariates], unit_fe, time_fe], axis=1)
    X_nounitfe = pd.concat([data[covariates], time_fe], axis=1)
    ## Primitives for training
    treatment_time = data.loc[(data[post]==1)][time_column].min()
    treated_units = data.loc[(data[treatment]==1)][unit_column].unique()
    treated_units = data[unit_column].isin(list(treated_units))
    pretreatment_time = data[time_column]<treatment_time
    
    ## Train a propensity score model on the treatment unit assignment
    p_fitted = propensity_model.fit(X,
                                    data[treatment].ravel())    
    ## Estimate the propensity score for all units
    p_hat = p_fitted.predict_proba(X)[:,1]
    
    ## Train an outcome model on the control units.
    y_control_fitted = outcome_model.fit(X.loc[ (treated_units==False) ],
                                    data.loc[(treated_units==False)][Y].ravel())    
    ## Predict the control units' outcomes
    mu_0 = y_control_fitted.predict(X)
        
    ## Construct the probability weight
    w_tr   = data[treatment].values
    w_yr   = data[treatment]/data[treatment].mean()
    w_contr = p_hat*(1-data[treatment])/(1-p_hat)
    w_contr /= w_contr.mean()
    weight = w_tr - w_contr

    
    y_11 = data.loc[(data[treatment]==1) & (data[time_column] >= treatment_time) ][Y]
    y_10 = data.loc[(data[treatment]==1) & (data[time_column] < treatment_time)  ][Y]
    y_01 = data.loc[(data[treatment]==0) & (data[time_column] >= treatment_time) ][Y]
    y_00 = data.loc[(data[treatment]==0) & (data[time_column] < treatment_time)  ][Y]    

    diff_treated = y_11.mean() - y_10.mean()
    
    weight_01 = weight[(data[treatment]==0) & (data[time_column] >= treatment_time) ]
    weight_00 = weight[(data[treatment]==0) & (data[time_column] < treatment_time) ]
    diff_control = (y_01*weight_01).mean() -  (y_00*weight_00).mean()
    
    diff_obs = diff_treated - diff_control    
    
    ## Estimate the counterfactual outcomes of treatment using the function trained on control
    mu_01 = mu_0[(data[treatment]==1) & (data[time_column] >= treatment_time) ]
    mu_00 = mu_0[(data[treatment]==1) & (data[time_column] >  treatment_time) ]
    diff_mu = mu_01.mean() - mu_00.mean()
        
    return {'atet':diff_obs - diff_mu, 'weight':weight, 
            'phat':p_hat,
            'diff_obs':diff_obs, 'diff_mu':diff_mu,
           'mu_0':mu_0}
```

## Comparison of the DRDID, TWFE, and IPW Models.
We want to conduct simulations of datasets and see over different simulations, which model give us the least amount of bias.

```python
at_gt = 5
twfe_est = []
ipw_est = []
dr_est = []

rfc = RandomForestClassifier(n_estimators=300, random_state=5247)
rfr = RandomForestRegressor( n_estimators=300, random_state=5247)

for r in range(50):
    df = synth_panel(seed_number = np.random.randint(0,2000), 
                    time_units = 20, 
                    x_units = 50, 
                    K_dim = 4,
                    at = at_gt )
    covariates_list = [x for x in df.columns if x[0]=='x']
    ## TWFE
    TWFE = DiD(df, 'Y','unitid','T',['treatmentxpost'])
    twfe_est.append(  TWFE.params[TWFE.params.index=='treatmentxpost'].values[0]  )

    ## IPW
    ipw = ipw_did(df, 'Y','unitid','T','treatment','treatmentxpost', 
           covariates_list, rfc )
    ipw_est.append(  ipw['atet']  )

    ## DRDiD
    dr = drdid(df, 'Y','unitid','T','treatment','treatmentxpost', 
           covariates_list, rfc,rfr )
    dr_est.append(  dr['atet']  )
```
