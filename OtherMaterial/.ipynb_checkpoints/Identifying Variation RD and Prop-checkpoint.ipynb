The purpose of this script is to show that treatment estimates are identified from a limited type of variation. Once you are predicting estimates outside that variation, you risk bias. We show two examples where we allow for heterogeneous treatment effects:
1. *Regression Discontinuity*: we allow the latent treatment effect to change for observations further away from the cutoff. If we were to only use the estimate at the cutoff and try to extrapolate, we find bias.
2. *Propensity Score Matching*: we illustrate this with a simple non-overlap sample. Treatment effects estimates under overlap are different the latent effects when there is no overlap.



```python
import numpy as np
import pandas as pd

%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import statsmodels.api as sm

```

```python
def winsorize(x=None, topcode=5, bottomcode=-5):
    x[x>=topcode]=topcode
    x[x<=bottomcode] = bottomcode
    return x

def dgp(N=1000,unobs=False,hte=None):
    ## Allow only one covariate for simplicity
    x = np.random.uniform(0.5,1,N)
    y = 2 + np.exp(x)/(1+np.exp(x))*5 + 5*x**3
    y += np.random.normal(0,0.25,N)

    
    ## Assign treatment.
    ## If we allow unobserved errors, treatment and y is based on an unobserved factors.
    ## In this case, we will truncate the propensity scores that are too low/high
    if unobs==True:
        zeta = np.random.uniform(-0.15,0.15,N)
        y += zeta
        latent_w = zeta + x
        latent_w = np.exp(latent_w) / (1+np.exp(latent_w))
#         latent_w[latent_w<0.25] = np.random.uniform(0.25,0.75, len(latent_w[latent_w<0.25]))
#         latent_w[latent_w>0.75] = np.random.uniform(0.25,0.75, len(latent_w[latent_w>0.75]))        
        w = (latent_w > np.percentile(latent_w,q=50)).astype(float)
    ## If we do not allow unobserved errors, use a cutoff rule in x
    else:
        w = (x >= 3).astype(float)
    

    
    ## Assign treatment effect and output the true counterfactuals
    w_te = hte(y=y,x=x,w=w)

    y0_true = y[:]
    y1_true = y + hte(y=y,x=x,w= np.ones(len(w)))
    
    
    
    return pd.DataFrame(data={'x1':x, 'y':y+w_te,'w':w,
                             'y_true0':y0_true, 'y_true1':y1_true})

def logit_hte(y=None,x=None,w=None):
    hte = 2*np.exp(x)/(1+np.exp(x))
    hte *=w
    return hte

def single_hte(y=None,x=None,w=None):
    return w*2

def linear_hte(y=None,x=None,w=None):
    return w*2*x

```

```python
df = dgp(N=2000, unobs=True,hte=linear_hte)
# df.describe()
display(df.groupby('w')['y'].mean())
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=(5,5))
ax.scatter(x=df.loc[df['w']==1]['x1'],
          y=df.loc[df['w']==1]['y'] ,
           alpha = 0.25,
           color='blue',
          label='Treat')
ax.scatter(x=df.loc[df['w']==0]['x1'],
          y=df.loc[df['w']==0]['y'],
           alpha = 0.25,
           color='orange',
          label='Control')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

Now we want to estimate the overall treatment effect. We will use two approaches:

1. **T-Learner** regression adjustment by predicting the outcome of treatment and control units. We will use random forest as the ML model here.
2. **(implicit) X-Learner** where we either explicitly or implicitly estimate a propensity score, study the overlap, and potentially narrow our focus. We will do a two-step approch estimating the propensity score and then OLS where we have overlap.

```python
'''
X-Learner
'''
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier


## Estimate a propensity score
rfc = RandomForestClassifier(random_state=78234, n_estimators=50)
rfc = GradientBoostingClassifier(random_state=78234, n_estimators=50)
prop_model = rfc.fit(df['x1'].array.reshape(-1,1), df['w'])
df['w_hat'] = prop_model.predict_proba(df['x1'].array.reshape(-1,1))[:,1]

## Plot the propensity scores and find where there is overlap
upper = np.percentile(df.loc[df['w']==0]['w_hat'], q=99)
lower = np.percentile(df.loc[df['w']==1]['w_hat'], q=1)

fig,ax = plt.subplots(ncols=1,nrows=1,figsize=(6,2.5))
et = ax.hist(df.loc[df['w']==1]['w_hat'], color='blue', alpha=0.25, label='Treatment')
ec = ax.hist(df.loc[df['w']==0]['w_hat'], color='orange', alpha=0.25, label='Control')
ax.vlines(x=[lower,upper],ymin=[0,0],ymax=[np.max([et[0].max() , ec[0].max()])]*2, color='black', label='Overlap')
ax.set_xlabel('Estimated Propensity Scores')
ax.legend()
plt.show()
display(df.groupby(df['w'])['w_hat'].describe())


df_overlap = df.loc[df['w_hat'].between(lower,upper)]
print('{0} in overlap, with {1:5.1f}% in treatment and {2:5.1f}% in control'.format(
    len(df_overlap),
    100*df_overlap['w_hat'].mean(),
    100*(1- df_overlap['w_hat'].mean())) )

## Just train on the overlap group
rfr = GradientBoostingRegressor(n_estimators=500, random_state=3823)

## Modeling Training and Estimate counterfactuals
rf_t = rfr.fit(df_overlap.loc[df['w']==1]['x1'].array.reshape(-1,1), df_overlap.loc[df_overlap['w']==1]['y'])
df['y_1x'] = rf_t.predict(df['x1'].array.reshape(-1,1))

rf_c = rfr.fit(df_overlap.loc[df['w']==0]['x1'].array.reshape(-1,1), df_overlap.loc[df_overlap['w']==0]['y'])
print( mean_squared_error(rf_t.predict(df['x1'].loc[df['w']==1].array.reshape(-1,1)), df.loc[df['w']==1]['y'])  ) 
df['y_0x'] = rf_c.predict(df['x1'].array.reshape(-1,1))

## Take the difference
df['x_learner_diff'] = df['y_1x'] - df['y_0x']


## Plot the results
fig,ax = plt.subplots(ncols=2,nrows=1,figsize=(15,5), sharex=True)
ax[0].scatter(x=df['x1'],
          y=df['y_1x'],
           alpha = 0.25,     
              color='blue',
          label='Estimated Treat')
ax[0].scatter(x=df['x1'],
          y=df['y_0x'],
           alpha = 0.25,
              color='orange',
          label='Estimated Control')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].legend()

ax[1].scatter(x = df['x1'],
             y = linear_hte(y=df['y'], x=df['x1'],w=np.ones(len(df))),
             color='magenta', alpha=0.25, label='Ground Truth HTE')
ax[1].scatter(x = df['x1'],
             y = df['x_learner_diff'],
             color='seagreen', alpha=0.25, label='Estimated HTE')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Estimated Treatment')
ax[1].legend()

```

sound-byte take-away.

Know where your natural experiment is taking place or risk getting the wrong answer.  Below I show that when there are treatment units that don't look like control units or vice versa (aka no overlap), any causal model will incorrectly extrapolate. Models that estimate a propensity score, like Propensity Score Matching and DML, can detect when you don't have overlap. Otherwise, this is impossible to detect when you face a lot of features and cannot easily compare treatment and control units, like with a T-Learner.
